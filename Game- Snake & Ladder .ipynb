```python
#the position of the snake and ladder from start to end point says the values

p_snakes = {17: 7, 54: 34, 62: 19, 98: 79} #snake bites- start:end
p_ladder = {3: 38, 24: 33, 42: 93, 72: 84} #ladder jump- start:end
```

```python
#function which let the player to reach the maximun value or come down by ladder(jump) or snake(bite)

def snake_ladder(player,player_init,dice_value):
    init=player_init  #assigning initial value to player_init
    player_init=player_init+dice_value  #adding the initial value with dice value
    
    #checking the player value less not 100 or not 
    if player_init>100: 
        print("Keep Trying you got ",player_init-100," more than 100 to win")
        return init
    
    #checking the player value to cross snake position gets bitten and reaches to which position.
    if player_init in p_snakes:
        final=p_snakes.get(player_init)
        
    #checking the else comndition if the player value cross the ladder position and its jumps and reaches to which position.
    elif player_init in p_ladder:
        final=p_ladder.get(player_init)
        
    #if those condition isn't satisfied it comes to else part and assign the player intial value as final.
    else:
        final=player_init

    return  final
```

```python
# Two-player game.
#this function check th players in which position are they and if one gets to final position,it win the match

def win(player,pos): #two arguments-the player and its position.
    if(pos==100):
        print("##### Congratulations "  + player +  " You won the match #####")
        print("##### Game Successfully Finsihed #####")
        return 1
    return 0
```
